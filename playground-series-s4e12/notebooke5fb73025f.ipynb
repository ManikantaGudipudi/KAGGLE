```python
train=pd.read_csv('/Users/manikantagudipudi/Desktop/Kaggle/playground-series-s4e12/train.csv')
test=pd.read_csv('/Users/manikantagudipudi/Desktop/Kaggle/playground-series-s4e12/test.csv')
train_y=train['Premium Amount']
train_x=train.drop('Premium Amount', axis=1)
```

```python
# Count NaN values per column
nan_per_column = train_x.isna().sum()

print("\nNumber of NaN values per column:\n")
print(nan_per_column)
```

```python
train_x = train_x.fillna(method='ffill')
train_y = train_y.fillna(method='ffill')
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Example dataset preparation
# Assuming train_x (features) and train_y (target) are already available
# Drop low mutual information features (optional)
selected_features = [
    'Health Score', 'Annual Income', 'Credit Score', 'Policy Start Date',
    'Age', 'Previous Claims', 'Vehicle Age', 'Occupation', 'Number of Dependents'
]
X = train_x[selected_features]
y = train_y  # Target: Principal Amount (continuous)

# Normalize the features
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# Train-test split
X_train, X_val, y_train, y_val = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Build the neural network model
model = Sequential([
    Dense(64, activation='relu', input_shape=(X_train.shape[1],)),  # First hidden layer
    Dense(32, activation='relu'),  # Second hidden layer
    Dense(16, activation='relu'),  # Third hidden layer
    Dense(1, activation='linear')  # Output layer for regression
])

# Compile the model
model.compile(optimizer='adam', loss='mse', metrics=['mae'])

# Train the model
history = model.fit(
    X_train, y_train,
    validation_data=(X_val, y_val),
    epochs=50,  # Adjust based on convergence
    batch_size=32,
    verbose=1
)

# Evaluate the model
loss, mae = model.evaluate(X_val, y_val, verbose=0)
print(f"Validation MAE: {mae:.4f}")

# Make predictions
predictions = model.predict(X_val)
```
